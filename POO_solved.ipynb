# 🚗  Principios de la Programación Orientada a Objetos (POO) en Python

La Programación Orientada a Objetos (POO) es un paradigma que organiza
el código en torno a **objetos**, que combinan datos (atributos) y
comportamiento (métodos).

En Python, los cuatro pilares fundamentales de la POO son:

-   Abstracción
-   Encapsulación
-   Herencia
-   Polimorfismo

------------------------------------------------------------------------

## 🔹 1. Abstracción

### 🧠 ¿Qué es?

La **abstracción** consiste en modelar entidades del mundo real
simplificando sus características, quedándonos solo con lo relevante.

------------------------------------------------------------------------

### 🚗 Ejemplo en Python

``` python
class Coche:
    def __init__(self, marca):
        self.marca = marca

    def arrancar(self):
        print("El coche arranca")

class CocheElectrico(Coche):
    def __init__(self, marca, bateria):
        super().__init__(marca)  # 👈 importante
        self.bateria = bateria

    def cargar(self):
        print("Cargando batería...")
```

------------------------------------------------------------------------

## 🔹 2. Encapsulación

### 🧠 ¿Qué es?

La **encapsulación** consiste en proteger los datos internos de un
objeto y controlar cómo se accede a ellos.

En Python: - No hay atributos realmente privados - Se usa `_atributo`
como convención

------------------------------------------------------------------------

### 🚗 Ejemplo en Python

``` python
class Coche:
    def __init__(self, marca, modelo):
        self.marca = marca
        self.modelo = modelo
        self._velocidad = 0

    def get_velocidad(self):
        return self._velocidad

    def set_velocidad(self, nueva_velocidad):
        if nueva_velocidad >= 0:
            self._velocidad = nueva_velocidad
```

------------------------------------------------------------------------

## 🔹 3. Herencia

### 🧠 ¿Qué es?

La **herencia** permite crear nuevas clases a partir de otras
existentes.

------------------------------------------------------------------------

### 🚗 Ejemplo en Python

``` python
class Coche:
    def arrancar(self):
        print("El coche arranca")

class CocheElectrico(Coche):
    def cargar(self):
        print("Cargando batería...")
```

------------------------------------------------------------------------

## 🔹 4. Polimorfismo

### 🧠 ¿Qué es?

El **polimorfismo** permite que diferentes objetos respondan al mismo
método de distintas maneras.

------------------------------------------------------------------------

### 🚗 Ejemplo en Python

``` python
class Coche:
    def arrancar(self):
        print("Arranca coche de combustión")

class CocheElectrico(Coche):
    def arrancar(self):
        print("Arranca coche eléctrico")

coches = [Coche(), CocheElectrico()]

for coche in coches:
    coche.arrancar()
```


# 🎮 Ejercicio: Combate entre personajes de videojuegos con POO

## Objetivo

Crea un pequeño sistema de combate por turnos usando Programación Orientada a Objetos en Python.

La idea es modelar personajes de videojuegos que puedan enfrentarse entre sí utilizando distintas acciones, como **atacar**, **defenderse** y **curarse**.

---

## Requisitos del ejercicio

### 1. Clase base `Personaje`

Crea una clase llamada `Personaje` que represente a un personaje genérico del juego.

Debe tener, al menos, estos atributos:

- `nombre`
- `salud`
- `ataque`
- `defensa`

Y los siguientes métodos:

- `atacar(objetivo)`: realiza un ataque sobre otro personaje
- `defenderse()`: aumenta temporalmente su defensa o activa un estado defensivo
- `curarse()`: recupera una cierta cantidad de salud
- `esta_vivo()`: devuelve `True` si el personaje sigue con vida y `False` en caso contrario

---

## 2. Crear dos personajes distintos

Crea dos personajes diferentes a partir de la clase `Personaje` o a partir de clases hijas.

Cada uno debe tener características distintas. Por ejemplo:

- uno con más ataque pero menos defensa
- otro con menos ataque pero más defensa

Puedes inspirarte en personajes de videojuegos, inventarles un nombre y asignarles estadísticas diferentes.

---

## 3. Combate por turnos

Programa un combate entre los dos personajes.

### Reglas del combate

- El personaje que empieza a atacar se decide de forma aleatoria.
- En cada iteración del bucle ataca un personaje.
- En la siguiente iteración ataca el otro.
- El combate termina cuando uno de los dos personajes se queda sin salud.
- Debe mostrarse por pantalla qué ocurre en cada turno:
  - quién ataca
  - cuánto daño hace
  - si el otro se defiende
  - si se cura --- ❗OJO ❗- Si añadís la opción de curar limitarla a un número determinado o podéis estar hasta mañana para ver quién gana.
  - cuánta salud le queda a cada personaje

---

## 4. Libertad para añadir complejidad

Puedes añadir toda la complejidad que quieras. Por ejemplo:

- que algunos personajes tengan habilidades especiales
- que defenderse reduzca el daño del siguiente ataque
- que curarse no se pueda hacer siempre
- que haya ataques críticos
- que se imprima un resumen final del combate

---

## 5. Qué se quiere practicar

Con este ejercicio se busca trabajar:

- clases y objetos
- herencia
- atributos y métodos
- interacción entre objetos
- lógica de control con bucles y condicionales
- diseño de un pequeño sistema orientado a objetos

---

## Pistas

- Puedes usar la librería `random` para decidir quién empieza.
- Un ataque puede calcularse como:
  `daño = ataque - defensa`
- Para evitar daños negativos, puedes usar:
  `max(1, daño)`
- Puedes usar un atributo booleano para indicar si un personaje está defendiendo.
- El combate puede ejecutarse con un bucle `while`.

---


````python
import random
from time import sleep
# Importamos sleep para que el bucle no se resuelva de inmediato
# Importamos random porque el enunciado pide que haya elementos aleatorios,
# como decidir quién empieza o introducir variación en el daño.
# Esto hace que el combate no sea totalmente predecible.


class Personaje:
    # Esta es la clase base o clase general.
    # En POO se usa una clase base cuando queremos definir aquello
    # que comparten todos los objetos de un mismo tipo.
    # En este caso, cualquier personaje del juego tendrá nombre,
    # salud, ataque, defensa y ciertas acciones comunes.

    def __init__(self, nombre, salud, ataque, defensa):
        # El constructor (__init__) sirve para inicializar el estado
        # de cada objeto cuando se crea.
        # Usamos atributos de instancia (self.algo) porque cada personaje
        # debe tener sus propios valores independientes.
        self.nombre = nombre
        self.salud = salud
        self.ataque = ataque
        self.defensa = defensa

        # Este atributo representa un estado del objeto.
        # Lo ponemos a False al principio porque, por defecto,
        # el personaje no está defendiendo.
        # En POO es habitual modelar estados internos con atributos.
        self.defendiendo = False

    def esta_vivo(self):
        # Este método encapsula una comprobación que necesitaremos
        # varias veces durante el combate.
        # En vez de escribir "self.salud > 0" en muchos sitios,
        # lo convertimos en comportamiento propio del objeto.
        # Esto mejora la legibilidad y respeta la idea de que
        # el objeto gestione su propio estado.
        return self.salud > 0

    def atacar(self, objetivo):
        # Este método modela una interacción entre objetos.
        # Un personaje no ataca "al vacío", sino a otro personaje,
        # por eso recibe un parámetro llamado objetivo.
        # Esto es un buen ejemplo de colaboración entre objetos en POO.

        # Añadimos una pequeña variación aleatoria al daño para que
        # el combate sea más dinámico y no repita siempre el mismo resultado.
        variacion = random.randint(-3, 3)

        # El daño base se calcula combinando el ataque del personaje actual
        # con la defensa del objetivo.
        # Accedemos a self.ataque porque es el que realiza la acción,
        # y a objetivo.defensa porque el daño depende del personaje que la recibe.
        dano = self.ataque + variacion - objetivo.defensa

        # Si el objetivo estaba defendiendo, reducimos aún más el daño.
        # Esto demuestra cómo el estado interno de un objeto (objetivo.defendiendo)
        # afecta al comportamiento del sistema.
        if objetivo.defendiendo:
            dano -= random.randint(3, 6)

        # Usamos max(1, dano) para garantizar que siempre se hace
        # al menos 1 punto de daño.
        # Esto evita valores negativos o cero, que podrían hacer
        # que el combate se alargara demasiado o fuese incoherente.
        dano = max(1, dano)

        # Restamos el daño a la salud del objetivo.
        objetivo.salud -= dano

        # Volvemos a usar max para evitar que la salud quede en negativo.
        # Esto ayuda a mantener el estado del objeto consistente.
        objetivo.salud = max(0, objetivo.salud)

        # Mostramos lo ocurrido en el turno.
        print(f"{self.nombre} ataca a {objetivo.nombre} y causa {dano} de daño.")
        print(f"Salud de {objetivo.nombre}: {objetivo.salud}")

        # Si el objetivo estaba defendiendo, informamos de ello
        # y desactivamos ese estado.
        # Lo hacemos porque la defensa en este diseño solo sirve
        # para el siguiente ataque recibido, no para todos.
        # Esto convierte la defensa en un efecto temporal.
        if objetivo.defendiendo:
            print(f"{objetivo.nombre} estaba defendiendo y redujo el daño.")
            objetivo.defendiendo = False

    def defenderse(self):
        # Este método cambia el estado interno del objeto.
        # En lugar de aumentar directamente la defensa numérica,
        # usamos un booleano para indicar que el personaje está en modo defensivo.
        # Esta solución es sencilla, clara y útil para un ejercicio de iniciación.
        self.defendiendo = True
        print(f"{self.nombre} se prepara para defenderse.")

    def elegir_accion(self):
        # Este método define cómo decide actuar un personaje.
        # Lo incluimos en la clase base para que cualquier personaje
        # tenga un comportamiento por defecto.
        #
        # Aquí aparece la idea de polimorfismo: todas las clases tendrán
        # este mismo método, pero cada una podrá redefinirlo a su manera.
        return random.choice(["atacar", "defenderse"])


# Clases hijas:
# Aquí aplicamos herencia.
# La idea es que Guerrero y Tanque son tipos concretos de Personaje:
# comparten el comportamiento general, pero tienen estadísticas distintas.
# Gracias a la herencia, evitamos repetir los métodos atacar, defenderse
# y esta_vivo, porque ya los heredan de la clase base.

class Guerrero(Personaje):
    def __init__(self, nombre):
        # Usamos super() para llamar al constructor de la clase padre.
        # Esto es importante en POO porque reutilizamos la lógica común
        # de inicialización en vez de reescribirla.
        # Solo cambiamos los valores concretos que definen a este tipo de personaje.
        super().__init__(nombre, salud=200, ataque=80, defensa=25)

    def elegir_accion(self):
        # Aquí sobrescribimos el método de la clase base.
        # Esto es polimorfismo: el mismo mensaje (elegir_accion)
        # provoca comportamientos distintos según el objeto concreto.
        #
        # El Guerrero tiene perfil ofensivo, así que tendrá más
        # probabilidad de atacar que de defenderse: 3 a 1.
        return random.choices(
            ["atacar", "defenderse"],
            weights=[3, 1]
        )[0]


class Tanque(Personaje):
    def __init__(self, nombre):
        # De nuevo usamos herencia para especializar un tipo de personaje.
        # El Tanque tiene más salud y defensa, pero menos ataque,
        # lo que permite diferenciar roles dentro del combate.
        super().__init__(nombre, salud=300, ataque=20, defensa=50)

    def elegir_accion(self):
        # También sobrescribimos el mismo método.
        # De nuevo hay polimorfismo: mismo método, distinta implementación.
        #
        # El Tanque tiene perfil defensivo, así que tendrá más
        # probabilidad de defenderse que de atacar: 3 a 1.
        return random.choices(
            ["atacar", "defenderse"],
            weights=[1, 3]
        )[0]


# Crear personajes
# Aquí instanciamos objetos concretos a partir de las clases.
# p1 y p2 son dos objetos diferentes con su propio estado.
p1 = Guerrero("Artorias")
p2 = Tanque("Smough")

print("===== PERSONAJES =====")
print(f"{p1.nombre} -> Salud: {p1.salud}, Ataque: {p1.ataque}, Defensa: {p1.defensa}")
print(f"{p2.nombre} -> Salud: {p2.salud}, Ataque: {p2.ataque}, Defensa: {p2.defensa}")

print("\n===== COMBATE =====")

# Elegir quién empieza
# El enunciado pide que el inicio sea aleatorio.
# En lugar de duplicar código, guardamos directamente en dos variables
# quién será atacante y quién defensor.
# Esto facilita mucho la lógica del combate posterior.
if random.randint(0, 1) == 0:
    atacante, defensor = p1, p2
else:
    atacante, defensor = p2, p1

print(f"Empieza {atacante.nombre}\n")

turno = 1

# El combate continúa mientras ambos personajes sigan vivos.
# Usamos el método esta_vivo() en lugar de comprobar directamente la salud
# porque así delegamos esa lógica al propio objeto.
while p1.esta_vivo() and p2.esta_vivo():
    print(f"--- Turno {turno} ---")

    # Ahora no comprobamos de qué clase es el personaje con isinstance.
    # En su lugar, le pedimos al propio objeto que decida qué hacer.
    #
    # Esto sí es polimorfismo: la llamada es la misma para cualquier personaje,
    # pero la acción devuelta depende de la clase concreta del objeto.
    accion = atacante.elegir_accion()

    # Según la acción elegida, el objeto ejecuta uno de sus métodos.
    # Esto refleja una idea central de la POO:
    # los objetos tienen datos (atributos) y comportamientos (métodos).
    if accion == "atacar":
        atacante.atacar(defensor)
    else:
        atacante.defenderse()

    print()

    # Tras la acción, comprobamos si el defensor ha sido derrotado.
    # Si ya no está vivo, termina el combate.
    if not defensor.esta_vivo():
        print(f"{defensor.nombre} ha sido derrotado.")
        print(f"🏆 Gana {atacante.nombre}")
        break

    sleep(0.25)

    # Intercambio de turnos
    # Intercambiamos los roles para que en la siguiente iteración
    # el defensor pase a ser atacante y viceversa.
    # Esta técnica evita tener dos bloques distintos de combate,
    # uno para cada personaje, y hace el código más elegante.
    atacante, defensor = defensor, atacante
    turno += 1
````

### Resumen:
- Encapsulación: cada personaje guarda su propio estado (salud, ataque, defensa, defendiendo).
- Métodos como comportamiento: en vez de manipular todo desde fuera, el propio objeto sabe atacar, defenderse y comprobar si está vivo.
- Herencia: Guerrero y Tanque reutilizan la clase Personaje y solo cambian sus estadísticas.
- Interacción entre objetos: un objeto Personaje puede actuar sobre otro mediante atacar(objetivo).
- Diseño reutilizable: si mañana quieres añadir Mago, Arquero o Curandero, puedes hacerlo heredando de Personaje.
- Polimorfismo: todos los personajes comparten el método elegir_accion(), pero cada clase lo implementa de forma distinta (por ejemplo, el Guerrero tiende a atacar y el Tanque a defenderse). El programa usa el mismo mensaje (atacante.elegir_accion()) sin necesidad de comprobar el tipo del objeto, y es cada instancia la que decide su comportamiento en tiempo de ejecución.
